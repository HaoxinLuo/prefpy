```python
from numpy import *

class Alternative:
    '''
        Class Alternative: Represents an alternative;
        Variables:
            qualities: a numpy.ndarray of the qualities of this alternative
            index: the integer index of this alternative in the GRUM's list, defaults to -1
    '''
    def __init__(self, qualities):
        self.qualities = qualities
        self.index = -1
        
class Agent:
    '''
        Class Agent: Represents a user;
        Variables:
            qualities: a numpy.ndarray of the qualities of this agent
            index: the integer index of this agent in the GRUM's list, defaults to -1
        Functions:
            utility: calculates the utility this agent views for a particular alternative
    '''
    def __init__(self, qualities):
        self.qualities = qualities
        self.index = -1
        
    def utility(self, alternative, interrelation):
        '''
            utility defined in (Xia et al) as: d_j + x_i * B * z_j^T + e_ij
            where:
                d_j is the intrinsic utility
                x_i is the vector of qualities about the agent
                B is the matrix of interrelation between agent and alternative qualities
                z_j is the vector of qualities about the alternative
                e_ij is the error term
        '''
        s1 = dot(self.qualities, interrelation.B)
        s2 = transpose(alternative.qualities)
        s3 = dot(s1, s2)
        s4 = s3 + interrelation.delta[alternative.index]
        
        return s4
        
        
class Theta:
    '''
        Class Interrelation: represents the utilities of alternatives as well as the
            relationship between qualities of agents and alternatives
        Variables:
            delta: the intrinsic utilities of an indexed list of  alternatives
            B: the matrix representing the interrelation
    '''
    def __init__(self, delta=[], B=[[]], theta=None, ):
        if(theta != None):
            self.delta = zeros(theta.delta.shape)
            self.B = zeros(theta.B.shape)
            return
        else:
            self.delta = delta
            self.B = B
        
    def __sub__(self, other):
        result = Theta(theta=other)
        
        for j in range(len(self.delta)):
            result.delta[j] = self.delta[j] - other.delta[j]
            
        for i in range(len(self.B)):
            for j in range(len(self.B[i])):
                result.B[i,j] = self.B[i,j] - other.B[i,j]
        
        
        return result
    
    def _print(self):
        print "\nDeltas: \n" , self.delta , "\nBs: \n" , self.B 
        
    def getConversion(self):
        numB = self.B.shape[0] * self.B.shape[1]
        numd = len(self.delta)
        t = numB + numd
        rVal = []
        
        index = 0
        for i in range(len(self.delta)):
            rVal[index] = self.delta[i]
            index += 1
        
        for i in range(len(self.B)):
            for j in range(len(self.B[i])):
                rVal[index] = self.B[i,j]
        
        return array(rVal)
        
class GRUM:
    '''
        Class GRUM: Represents a General Randomized Utility Model
        Overview: 
            The GRUM is in two primary states: unlocked and locked. In the unlocked state, new agents and 
            alternatives can be added to the GRUM. The purpose of locking is to prevent additional information 
            from coming into the class that may otherwise affect dimensions of calculated matrices. 
            In the locked state is when learning algorithms can be applied, moving one state of the GRUM to
            another.
        Variables:
            Agents: an array of agents
            Alternatives: an array of alternatives
            Theta: an interrelation matrix and a list of utilities
            locked: a boolean telling the state of the GRUM
            LatentSpace: ndarray containing the utilites of all agents for all alternatives
        Functions:
            pushAgent: adds an agent to the list of agents, pairing its index in the agent itself
            pushAlternative: same as pushAdent, except for alternatives
            lock: locks the class
            makeLatentSpace: generates a variable LatentSpace that contains the utilities of all agents for all alternatives
    '''
    def __init__(self):
        self.Agents = []
        self.Alternatives = []
        self.Theta_Initial = Theta()
        self.locked = False
        
    def pushAgent(self, agent):
        if(not self.locked):
            agent.index = len(self.Agents)
            self.Agents.append(agent)
        
    def pushAlternative(self, alternative):
        if(not self.locked):
            alternative.index = len(self.Alternatives)
            self.Alternatives.append(alternative)
            
    def lock(self):
        self.locked = True
        
        
    def _getLatentSpace(self, theta):
        shape = (len(self.Agents) , len(self.Alternatives))
        LatentSpace = empty(shape)
        
        for i in range(len(self.Agents)):
            for j in range(len(self.Alternatives)):
                LatentSpace[i,j] = self.Agents[i].utility(self.Alternatives[j],theta)
        return LatentSpace
    
    def _getEtaVals(self, LatentSpace, Eta=(lambda x: x)):
        '''
            Returns a matrix representing the values of Eta applied to the LatentSpace.
            This is a horribly inefficient way to do this (memory wise) 
        '''
        EtaArr = empty(LatentSpace.shape)
        for i in range(len(LatentSpace)):
            for j in range(len(LatentSpace[i])):
                EtaArr[i,j] = Eta(LatentSpace[i,j])
                
    def _getAVals(self, EtaVals, A=(lambda x: x)):
        # TODO: Implement A Func Matrix generation
        return -1
        
    def _mStep(self, theta, thetaT, S, eta, A):
        L1 = self._getLatentSpace(theta)
        L2 = self._getLatentSpace(thetaT)
        
        Eta1 = self._getEtaVals(L1, eta)
        Eta2 = self._getEtaVals(L2, eta)
        
        #TODO: Multiply Eta1,2 by S_ij^t
        
        A1 = self._getAVals(Eta1, A)
        A2 = self._getAVals(Eta2, A)
        
        EtaD1 = self._getDTheta(Eta1, Eta2, theta, thetaT)
        EtaD2 = self._getHess()
        
        AD1 = self._getDTheta(A1, A2, theta, thetaT)
        AD2 = self._getHess()
        
        sum1 = Theta(theta)
        sum2 = Theta(theta)
        
        for i in range(len(Eta1)):
            for j in range(len(Eta1[i])):
                sum1 += EtaD1[i,j] - AD1[i,j]
                sum2 += EtaD2[i,j] - AD2[i,j]
                
        #TODO: Correct output and return value
        
        
    def _getHess(self, args):
        # TODO: Implement Hessian Matrix generation
        return -1
        
    def _getDTheta(self, matrix1, matrix2, theta1, theta2):
        '''
            matrix1,2 are m x n matrices 
            theta1,2 are theta and theta^t respectively
            rval is a new theta-like shape
        '''
        rval = Theta(theta = theta1)
        
        for i in range(len(matrix1)):
            for j in range(len(matrix1[i])):
                dMat = matrix2[i,j] - matrix1[i,j]
                dTheta = theta2 - theta1
                
                rval.delta[j] += dMat/ dTheta.delta[j]
                for m in range(len(dTheta.B)):
                    for n in range(len(dTheta.B[m])):
                        rval.B[m,n] += dMat/ dTheta.B[m,n]
                
        return rval
        
    def M(self, theta, theta_t, S, etaFunc, AFunc):
        thetaArr = theta.getConversion()
        thetaTArr = theta_t.getConversion()
        p = len(thetaArr)
        lat1 = self._getLatentSpace(theta)
        lat2 = self._getLatentSpace(theta_t)
        first = zeros((1, p))
        hess = zeros((p, p))
        
        for i in range(len(self.Agents)):
            for j in range(len(self.Alternatives)):
                eta1 = etaFunc(lat1[i,j])
                eta2 = etaFunc(lat2[i,j])
                A1   = AFunc(eta1)
                A2   = AFunc(eta2)
                
                for x in range(p):
                    
                    first[x] += ((eta2 - eta1) * S) / (thetaTArr[x] - thetaArr[x])
                    first[x] -= (A2 - A1) / (ThetaTArr[x] - ThetaArr[x])
                    
                    for y in range(p):
                        hess[x,y] += (((eta2 - eta1)**2) * S) / ((thetaTArr[x] - thetaArr[x]) * (thetaTArr[y] - thetaArr[y]))
                        hess[x,y] -= ((A2 - A1)**2)  / ((thetaTArr[x] - thetaArr[x]) * (thetaTArr[y] - thetaArr[y]))
                        
        invhess = linalg.matrix_power(hess, -1)
        result = dot(invhess, first)
        rtheta = Theta(theta)
        
        index = 0
        
        for elem in rtheta.delta:
            elem = result[index]
            index += 1
        
        for row in rtheta.B:
            for elem in row:
                elem = result[index]
                index += 1
                
        return theta_t - rtheta
            
        

```
